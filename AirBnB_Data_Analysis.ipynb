### 1. Analyze Change in Price

```python
data['month'] = pd.to_datetime(data['month'], format='%Y_%m')
```

```python
average_price_by_month = data.groupby('month')['price'].mean()

plt.figure(figsize=(10, 6))
plt.plot(average_price_by_month, marker='o', linestyle='-')
plt.title('Average Airbnb Listing Price Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

We see an increase in the average price post pandemic, we see a significant spike in the graph. Even thought the prices fluctuate between 2020-2021, the overall jump is still higher than the pre-apandemic era. This is indicative of the fact that AirBnb prices rose during the pandemic.

### 2. Analyze Change in room type

```python
data['year'] = data['month'].dt.year

room_type_counts_yearly = data.groupby(['year', 'room_type']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='count', hue='room_type', data=room_type_counts_yearly, palette='pastel')

for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', fontsize=8, color='black', xytext=(0, 5),
                       textcoords='offset points')


plt.title('Room Type Distribution Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Room Type')
plt.show()
```

Analyzing the room type distribution per year shows that the the trend in listing entire home/apartments and private roons made a significant jump starting 2018. Shared rooms made a jump as well but not as significant. As would be expected, when the pandemic hit in 2020, the listings for each type reduced, except for hotel rooms. We see a small rise in the number - this could probably be attributed to hotel rooms being used for isolation. Although the rise isn't significant, it is a hypothesis derived from the graph. Post pandemic, there is a sharp drop in overall listings. Infact, we can observe that it is the lowest since 2015. Therefore, we can arrive at the conclusion that the number of listings overall, were significantly impacted.

```python
pre_pandemic = data[data['month'] < '2020-03']
room_type_distribution_pre = pre_pandemic['room_type'].value_counts()

post_pandemic = data[data['month'] >= '2020-03']
room_type_distribution = post_pandemic['room_type'].value_counts()
sns.set_palette("pastel")


plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.pie(room_type_distribution_pre, labels=room_type_distribution_pre.index, autopct='%1.1f%%', startangle=140)
plt.title('Room Type Distribution Before COVID-19')

plt.subplot(1, 2, 2)
plt.pie(room_type_distribution, labels=room_type_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Room Type Distribution During COVID-19')

plt.show()
```

The room distribution based on the type tells us that the overall distribution remains more or less the same with slight changes in the percentages individually. There's a small rise in the entire homes/apartments category, as well as hotel rooms. Private rooms and shared rooms have a slight decline. This could indicate how due to covid isolation protoicols, and safety measures, listing out entire units became more common, because of individuals not preferring to share a room or even the same apartment with strangers.

### 3. Analyze Availability Trend

```python
availability_by_month = data.groupby('year')['availability_365'].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=availability_by_month.index, y=availability_by_month.values, marker='o')
plt.title('Average Availability Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Average Availability (365 days)')
plt.show()
```

Availability trends can be seen declining. As listings plummet, it will also impact the avaiability which can be seen from the figure.

### 4. Analyze Host Listings

```python
host_listings_count_by_month = data.groupby('year')['calculated_host_listings_count'].sum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=host_listings_count_by_month.index, y=host_listings_count_by_month.values, marker='o')
plt.title('Total Host Listings Count Over Time')
plt.xlabel('Month')
plt.ylabel('Total Host Listings Count')
plt.show()
```

We observe that overall host listings plummeted during the pandemic. This goes hand in hand with the analysis of the yearly room type distribution.

### 5. Neighbourhood Analysis

```python
data
```

```python
pre_pandemic_data = data[data['month'] < '2020-03-01']

top_10_neighborhoods_before_covid = pre_pandemic_data.groupby('neighbourhood')['price'].mean().nlargest(10)

data['is_covid'] = data['month'] >= '2020-03-01'

after_covid_data = data[data['is_covid']]

top_10_neighborhoods_after_covid = after_covid_data.groupby('neighbourhood')['price'].mean().nlargest(10)

plt.figure(figsize=(15, 8))

# Before COVID
plt.bar(top_10_neighborhoods_before_covid.index, top_10_neighborhoods_before_covid.values, color='red', label='Before COVID')

# After COVID
plt.bar(top_10_neighborhoods_after_covid.index, top_10_neighborhoods_after_covid.values, color='pink', label='After COVID', alpha=0.6, hatch='/', edgecolor='black')

plt.title('Top 10 Neighborhoods by Average Price Before and After COVID')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
